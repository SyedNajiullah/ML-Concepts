```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import load_digits
```

# Gamma Hadron particle classification

```python
columns_names = ["fLength", "fWidth", "fSize", "fConc", "fConc1", "fAsym", "fM3Long", "fM3Trans", "fAlpha", "fDist", "class"]
df = pd.read_csv("/content/magic04.data", names =columns_names)

df.head()
```

```python
print(df["class"].unique())
df["class"] = (df["class"] == 'g').astype(int)
print(df["class"].unique())
```

```python
X = df[df.columns[:-1]]
y = df[df.columns[-1]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Logistic regression model

```python
lr_model = LogisticRegression().fit(X_train, y_train)
pre = lr_model.predict(X_test)

print(classification_report(y_test, pre))
```

# MNIST data classification

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
for i in range(5): # printing first 5 images
  print("-"*5)
  print("Numeric data")
  print("-"*5)
  print(digits.data[i])
  print("-"*5)
  print("Number")
  print("-"*5)
  print(digits.target[i])
  print("-"*5)
  print("Image")
  print("-"*5)
  plt.gray()
  plt.matshow(digits.images[i])
  plt.show()
```

```python
X = digits.data
y = digits.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr_model = LogisticRegression().fit(X_train, y_train)
pre = lr_model.predict(X_test)
print(pre)
```

```python
print(classification_report(y_test, pre))
```

Predicting 6 = which is digits.images[67]

```python
plt.matshow(digits.images[67])
```

```python
lr_model.predict([digits.data[67]]) # accurately predicted 6
```

```python
cm = confusion_matrix(y_test, pre)
sns.heatmap(cm, annot=True)
plt.xlabel("True Lable")
plt.ylabel("Predicted label")
plt.title("Confusion Matrix")
plt.show()
```
