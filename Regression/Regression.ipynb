# Linear Regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("/content/HousePrice.xlsx")
df
```

```python
X = df[["Area"]]
y = df["Price(Dollars)"]
```

```python
model = LinearRegression().fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[3.00]])
```

# Multiple Linear Regression

```python
df = pd.read_excel("/content/SalaryPrediction.xlsx")
df
```

```python
median_test_score = df['TestScore'].median()
df['TestScore'].fillna(median_test_score, inplace=True)
df
```

```python
X = df[df.columns[:-1]]
y = df[df.columns[-1]]
```

```python
model = LinearRegression().fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[4, 10, 10]]) # candidate with 4years experience, 10 Test score and 10 Interview score will have 70446.37817498 salary
```
