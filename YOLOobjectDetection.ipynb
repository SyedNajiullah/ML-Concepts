# Image to detect

```python
model = YOLO("yolo11n.pt")  # small model, fast for testing
```

```python
results = model("/content/Screenshot 2025-11-18 115115.png")
```

```python
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    result.show()
    result.save(filename="result.jpg")  # save to disk
```
