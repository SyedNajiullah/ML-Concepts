Scatter plot on x= flipper length and y = body mass gropued by island

Scatter plot on x= flipper length and y = body mass gropued by sex

Male have the highest bodymass

```python
# Scatter Plot on flipper_length_mm vs body_mass_g grouped by gender
sns.set_style("ticks") # allowed styles are white, dark, whitegrid, darkgrid, ticks
sns.set_context("notebook") # allowed context are paper, notebook, talk, poster
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue = "sex")
sns.despine()
```

```python
sns.scatterplot(data=df, x="species", y="body_mass_g", hue = "island", style = "sex", alpha=0.7)
```

Adelie have all ind of species penguins, dream island have chinstrap, and Biscoe island have Gentoo specie only

```python
sns.stripplot(data=df, x="species", y="body_mass_g", hue = "island", dodge = True)
```

Larger bodymass of penguinsare found in Gentoo

```python
sns.set_context("notebook")
sns.swarmplot(data=df, x="species", y="body_mass_g", hue = "island")
```

Male have higher bodymass

```python
sns.histplot(data=df, x="body_mass_g", hue="sex", multiple="stack")
```

```python
sns.regplot(data=df, x="body_mass_g", y="flipper_length_mm", color="green")
```

```python
sns.lineplot(data=df, x="body_mass_g", y="flipper_length_mm", hue="island", style="sex")
```

With respect to females make have higher body mass

```python
sns.barplot(data=df, x="species", y="body_mass_g", hue="sex")
```

```python
sns.countplot(data=df, x="species", hue="island")
```

Confusion matric between all numeric values of the dataset

```python
columns = ["bill_length_mm", "bill_depth_mm",	"flipper_length_mm",	"body_mass_g"]
print(df[columns].corr())
sns.heatmap(data=df[columns].corr(), annot=True, cmap="Blues", vmin=-0.2)
```

Plot for all numeric values

```python
columns = ["bill_length_mm", "bill_depth_mm",	"flipper_length_mm",	"body_mass_g"]
sns.pairplot(data=df[columns], palette="Set2")
```
