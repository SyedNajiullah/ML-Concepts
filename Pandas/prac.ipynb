```python
df1 = pd.DataFrame({
    "city": ["lahore", "islamabad", "karachi", "multan"],
    "temperature": [20, 13, 45, 36]
})
df2 = pd.DataFrame({
    "city": ["islamabad", "new york"],
    "humidity": [65, 68]
})
print(df1)
print(df2)
```

```python
pd.merge(df1, df2, on="city") # common
```

```python
pd.merge(df1, df2, on="city", how="outer") # outer
```

```python
pd.merge(df1, df2, on="city", how="left") # left
```

```python
pd.merge(df1, df2, on="city", how="right") # right
```

```python
pd.merge(df1, df2, on="city", how="outer", indicator=True) # union
```

```python
df1 = pd.DataFrame({
    "city": ["lahore", "islamabad", "karachi", "multan"],
    "temperature": [20, 13, 40, 20],
    "humidity": [65, 68, 75, 70]
})
df2 = pd.DataFrame({
    "city": ["canbra", "havana", "brazilia", "london"],
    "temperature": [20, 13, 40, 20],
    "humidity": [65, 68, 75, 70]
})
```

```python
pd.merge(df1, df2, on="city", how="outer")
```

```python
pd.merge(df1, df2, on="city", how="outer", suffixes=('_first', '_second'))
```
