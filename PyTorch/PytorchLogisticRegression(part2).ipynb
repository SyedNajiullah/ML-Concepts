```python
import numpy as np
import torch
import torch.nn as nn
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
iris_dataset = datasets.load_iris()
dir(iris_dataset)
```

```python
X, y = iris_dataset.data, iris_dataset.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.int64)) # Use int64 for target variables
y_test = torch.from_numpy(y_test.astype(np.int64))   # Use int64 for target variables
```

```python
class LogisticRegression(nn.Module):
  def __init__(self, input_dim, output_dim):
    super(LogisticRegression, self).__init__()
    self.linear = nn.Linear(input_dim, output_dim)
  def forward(self, x):
    return self.linear(x)
```

```python
model = LogisticRegression(4, 3)
loss = nn.CrossEntropyLoss()
optimiser = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(99999):
  y_pred = model(X_train)
  l = loss(y_pred, y_train)
  l.backward()
  optimiser.step()
  optimiser.zero_grad()
  print(f"Epoch: {epoch}, Loss: {l}")
```

```python
with torch.no_grad():
  y_predicted = model(X_test)
  y_predicted_cls = torch.argmax(y_predicted, dim=1)
  acc = y_predicted_cls.eq(y_test).sum() / float(y_test.shape[0])
  print(f"Accuracy: {acc.item()*100}")
```
