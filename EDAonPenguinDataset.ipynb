Histogram for each graph

```python
columns_names = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g", "sex"]
for label in columns_names[:-1]:
  plt.hist(df[df["sex"] == "Male"][label], color="blue", alpha=0.7, label = "male", density = True)
  plt.hist(df[df["sex"] == "Female"][label], color="red", alpha=0.7, label = "female", density = True)

  plt.title(label)
  plt.ylabel("Probability")
  plt.xlabel(label)
  plt.legend()
  plt.show()
```
