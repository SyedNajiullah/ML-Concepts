```python
import numpy as np
import PIL.Image as Image
import tensorflow as tf
import tensorflow_hub as hub
```

```python
IMAGE_SHAPE = (224, 224)
classifier = hub.KerasLayer(
    "https://tfhub.dev/google/tf2-preview/mobilenet_v2/classification/4",
    input_shape=IMAGE_SHAPE+(3,),
)
```

```python
goldfish_img = Image.open("/content/goldfish.png")
np.array(goldfish_img).shape
```

```python
goldfish_img = goldfish_img.resize(IMAGE_SHAPE)
np.array(goldfish_img).shape
```

```python
goldfish_img_rgb = goldfish_img.convert('RGB')
goldfish_np = np.array(goldfish_img_rgb)
goldfish_np = goldfish_np / 255.0
goldfish_batch = goldfish_np[np.newaxis, ...]
predictions = classifier(goldfish_batch)
print("Predictions shape:", predictions.shape)
print("Prediction: ", np.argmax(predictions)) # index 2 we have goldfist
```

```python
image_labels = []
with open("/content/ImageNetLabels.txt", "r") as f:
    image_labels = f.read().splitlines()
print("Final prediction: ", image_labels[np.argmax(predictions)])
```
