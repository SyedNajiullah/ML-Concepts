```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

# Salary Prediction

```python
df = pd.read_excel("/content/SalaryPrediction.xlsx")
df.head()
```

```python
company_encoder = LabelEncoder()
job_encoder = LabelEncoder()
degree_encoder = LabelEncoder()
```

```python
inputs = df.drop("sallary_more_then_100k", axis="columns")
target = df["sallary_more_then_100k"]
```

```python
inputs["company_n"] = company_encoder.fit_transform(inputs["company"])
inputs["job_n"] = company_encoder.fit_transform(inputs["job"])
inputs["degree_n"] = company_encoder.fit_transform(inputs["degree"])
inputs.head()
```

```python
inputs.drop(["company", "job", "degree"], axis="columns", inplace=True)
inputs.head()
```

```python
dtc = DecisionTreeClassifier().fit(inputs, target)
dtc.score(inputs, target)
```

```python
dtc.predict([[2, 2, 0]]) # google, sales, bachelors
```

```python
dtc.predict([[2, 0, 1]]) # google, business, masters
```

# Play FootBall prediction

```python
df = pd.read_excel("/content/PlayFootballPrediction.xlsx")
df.head()
```

```python
le = LabelEncoder()
df["Weather"] = le.fit_transform(df["Weather"])
df["Temperature"] = le.fit_transform(df["Temperature"])
df["Humidity"] = le.fit_transform(df["Humidity"])
df["Wind"] = le.fit_transform(df["Wind"])
df["Play Football"] = (df["Play Football"] == "Yes").astype(int)
df.head()
```

```python
X = df.drop("Play Football", axis="columns")
y = df["Play Football"]
```

```python
dtc = DecisionTreeClassifier().fit(X, y)
dtc.score(X, y)
```

```python
dtc.predict([[1, 3, 1, 0, 1]])
```

```python
dtc.predict([[2, 3, 1, 0, 0]])
```
