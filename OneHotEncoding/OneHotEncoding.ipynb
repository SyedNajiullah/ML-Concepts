```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_excel("/content/housing.xlsx")
dflabel = df
df.head()
```

```python
dummies = pd.get_dummies(df.town)
dummies["monroe township"] = (dummies["monroe township"] == True).astype(int)
dummies["robinsville"] = (dummies["robinsville"] == True).astype(int)
dummies["west windsor"] = (dummies["west windsor"] == True).astype(int)
dummies
```

```python
dataframe = pd.concat([df, dummies], axis="columns")
dataframe
```

```python
df = dataframe.drop(["town", "west windsor"], axis="columns")
df
```

```python
X = df.drop("price", axis="columns")
y = df["price"]
```

```python
model = LinearRegression().fit(X, y)
prediction = model.predict([[3100, 0, 1]])
print(prediction)
print(model.score(X, y))
```
