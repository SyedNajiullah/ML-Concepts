Importing

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

Creating Sample Data

```python
data = {
    'Area_sqft': [2000, 1500, 2500, 1800, 3000],
    'Built_Year': [2010, 2015, 2005, 2020, 2012],
    'Location': ['Urban', 'Suburban', 'Urban', 'Rural', 'Suburban'],
    'Price': [250000, 180000, 300000, 150000, 320000]
}
df = pd.DataFrame(data)
df
```

Creating new feaures

1. House Age
2. Price per square feet
3. Location encoder
4. Scaling (Standard scaling)

Droping unnecessary columns

```python
current_year = 2025
df['House_Age'] = current_year - df['Built_Year']   # new numeric feature
df['Price_per_sqft'] = df['Price'] / df['Area_sqft']  # derived feature
```

```python
label_encoder = LabelEncoder()
df['Location_Encoded'] = label_encoder.fit_transform(df['Location'])
```

```python
scaler = StandardScaler()
df[['Area_sqft', 'House_Age', 'Price_per_sqft']] = scaler.fit_transform(df[['Area_sqft', 'House_Age', 'Price_per_sqft']])
```

```python
df.drop(columns=['Built_Year', 'Location'], axis=1, inplace=True)
```
