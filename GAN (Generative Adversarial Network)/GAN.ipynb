```python
import numpy as np
import matplotlib.pyplot as plt

def display_images_grid(data_matrix, indices, rows, cols):
    if any(i >= data_matrix.shape[0] or i < 0 for i in indices):
        print("One or more indices are out of range.")
        return

    fig, axes = plt.subplots(rows, cols, figsize=(cols * 2, rows * 2))
    fig.subplots_adjust(hspace=0.5, wspace=0.1)

    for ax, index in zip(axes.flatten(), indices):
        # Check if the current index exists in the indices list
        if index < len(indices):
            # Reshape the 1D array (row of the data matrix) back to 2D (assuming 100x100)
            image = data_matrix[index].reshape(28,28)
            ax.imshow(image, cmap='gray')
            ax.set_title(f"Image at index {index}")
            ax.axis('off')  # Turn off axis numbers and ticks
        else:
            ax.axis('off')  # Hide empty subplots

    plt.show()

# Example usage
# Assuming `data_matrix` is already loaded with image data
# Display the first 10 images in a 2x5 grid
indices_to_display = list(range(20))  # You can specify any range of indices here
display_images_grid(data_matrix, indices_to_display, 4, 5)
```

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, Model

# Setting up pandas to display all columns
pd.set_option('display.max_columns', None)

# Load dataset
def load_data():
    df_train = pd.read_csv('/content/train.csv')
    data_matrix = df_train.drop('label', axis=1).values
    data_matrix = ((data_matrix.astype(np.float32) / 255.0) * 2 - 1.0)
    return data_matrix, df_train['label']

data_matrix, labels = load_data()



# Define the generator
def build_generator(input_noise_dim=100, output_dim=784):
    model = models.Sequential([
        layers.Dense(100, use_bias=False, input_shape=(input_noise_dim,)),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.Dense(100),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.Dense(output_dim, activation='tanh')
    ])
    return model

# Define the discriminator
def build_discriminator(input_dim=784):
    model = models.Sequential([
        layers.Dense(512, input_shape=(input_dim,)),
        layers.LeakyReLU(0.01),
        layers.Dropout(0.3),
        layers.Dense(256),
        layers.LeakyReLU(0.01),
        layers.Dropout(0.3),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(0.005, 0.5), metrics=['accuracy'])
    return model

# Define the GAN
def build_gan(generator, discriminator):
    discriminator.trainable = False
    model = models.Sequential([generator, discriminator])
    model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(0.005, 0.5), metrics=['accuracy'])
    return model

# Instantiate the models
generator = build_generator()
discriminator = build_discriminator()
gan = build_gan(generator, discriminator)

# Display models
generator.summary()
discriminator.summary()
gan.summary()

# Training function
def train_gan(generator, discriminator, gan, data_matrix, iterations=100, batch_size=128):
    for iteration in range(iterations):
        idx = np.random.randint(0, data_matrix.shape[0], batch_size)
        real_images = data_matrix[idx]

        noise = np.random.normal(0, 1, (batch_size, 100))
        generated_images = generator.predict(noise)

        # Train Discriminator
        discriminator.trainable = True
        d_loss_real = discriminator.train_on_batch(real_images, np.ones((batch_size, 1)))
        d_loss_fake = discriminator.train_on_batch(generated_images, np.zeros((batch_size, 1)))
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Train Generator
        discriminator.trainable = False
        noise = np.random.normal(0, 1, (batch_size, 100))
        g_loss = gan.train_on_batch(noise, np.ones((batch_size, 1)))

        # Print progress
        print(f"Iteration {iteration + 1}/{iterations} - D Loss: {d_loss[0]}, D Accuracy: {d_loss[1]}, G Loss: {g_loss[0]}, G Accuracy: {g_loss[1]}")
        if iteration % 20 == 0:
            display_generated_images(generator)


# Display generated images
def display_generated_images(generator, num_examples=10):
    noise = np.random.normal(0, 1, (num_examples, 100))
    generated_images = generator.predict(noise)
    plt.figure(figsize=(10, 1))
    for i in range(num_examples):
        plt.subplot(1, 10, i+1)
        plt.imshow(generated_images[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    plt.show()

# Training the GAN
train_gan(generator, discriminator, gan, data_matrix, iterations=800, batch_size=128)

# Displaying generated images
display_generated_images(generator)
```
